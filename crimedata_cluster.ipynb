# K Means

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
data1 = pd.read_csv("C:/Users/pranay/Downloads/crime_data.csv")
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(data1.iloc[:,1:])
df_norm
```

```python
# Elbow curv



wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
data1['Clust']=x
```

```python
data1
```

```python
data1.iloc[:,1:5].groupby(data1.Clust).mean()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data2 = pd.read_csv("C:/Users/pranay/Downloads/crime_data.csv")
```

```python
df=data2.iloc[:,1:5]
```

```python
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
pd.concat([data2,cl],axis=1)
```
